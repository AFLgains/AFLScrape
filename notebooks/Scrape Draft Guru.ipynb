In this notebook, we'll demonstrate how to retrieve data automatically from draft guru

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from collections import ChainMap
import math
import numpy as np
```

```python
def make_soup(website):
    if 'https' in website:
        page = requests.get(website, timeout=120)
    else:
        page = requests.get('https://' + website, timeout=120)
    soup = BeautifulSoup(page.content, 'html.parser')
    page.close()
    return soup
```

```python
TEAM = "geelong"
```

```python
URL = "https://www.draftguru.com.au/clubs/"+TEAM
```

```python
soup = make_soup(URL)
```

```python
[(a.get('class'),a.text) for a in soup.find_all('td')]
```
